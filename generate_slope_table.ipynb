```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Differential equation
y_prime = lambda x, y: x**2 - y - 2

# Define bound
x = np.linspace(-6,6,13)
y = np.linspace(-6,6,13)

# Store result
result = np.zeros(shape = (len(x),len(y)))
```

```python
for i in range(len(x)):
    for j in range(len(y)):
        result[i][j] = y_prime(x[i], y[j])
        
result = np.transpose(result)
result = pd.DataFrame(result, index = y)
result.columns = x
result
```
